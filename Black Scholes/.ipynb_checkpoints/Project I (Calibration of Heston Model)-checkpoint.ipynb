### Heston model
eqns, charcteristic function

Once we have the implied volatility, we will try to optimize the call price in **Heston Model**, using optimization techniques

Steps:
1. Use FFT technique to calculate the model call price c_m
2. Calculate the objective function H(p)

p consistes of the parameters of heston model{$\sigma,\theta,\rho,\kappa,v_0$}

3. Optimize the objective function using optimization techniques


### **Local Objective Function**

The **local objective function** \( H(p) \) can be defined as `absolute error`

$$
H(p) = w_i \left(| c_i(K, T) - c_{m_i}(K, T, p)| \right)
$$


### **Global Objective Function**

The **global objective function** \( H(p) \) is defined as `root mean square error` 

$$
H(p) =\sqrt{ \frac{\sum_{i} w_i \left( c_i(K, T) - c_{m_i}(K, T, p) \right)^2}{n}}
$$

Where:

- $ p = \{\sigma, \theta, \rho, \kappa, v_0\} $ are the Heston model parameters:
    - $ \sigma $: volatility of variance
    - $ \theta $: long-term variance
    - $ \rho $: correlation between asset price and variance
    - $ \kappa $: speed of mean reversion
    - $ v_0 $: initial variance
    
- $ c_i(K, T) $: observed market call price for strike \( K \) and maturity \( T \).
- $ c_{m_i}(K, T, p) $: model call price computed using FFT for \( K \), \( T \), and parameters \( p \).
- $w_i $: weight assigned to each observation (typically set to 1).
- $n$ is the number of data points.


### Calculating the Model price using closed form and Local Optimization

My goal is to fit the model locally to each combination of strike and maturity, optimizing parameters individually for each pair. This approach is often used when:

    1. You aim to model specific market segments or conditions.

    2. You want to capture the dynamics of all parameters {$\sigma,\theta,\rho,\kappa,v_0$} in Heston model
    for each strike and maturity separately.

**Steps**

- We will use a range of values for kappa, theta, rho, sigma, v0.
- Then we will optimize it for each market price, strike and Maturity from data.
- We will do it for all the rows of data
- add optimized kappa, theta, rho, sigma, v0 as columns in data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
from scipy.stats import norm

def heston_call_price(S0, K, T, r, q, kappa, theta, sigma, rho, v0):
    """
    Simple closed-form approximation for Heston call option price.
    """
    d1 = (np.log(S0 / K) + (r - q + 0.5 * v0) * T) / (np.sqrt(v0 * T))
    d2 = d1 - np.sqrt(v0 * T)
    
    call_price = (S0 * np.exp(-q * T) * norm.cdf(d1) - 
                  K * np.exp(-r * T) * norm.cdf(d2))
    
    return call_price

def objective_function(params, S0, K, T, r, q, market_price):
    """
    Objective function to minimize the difference between model and market prices.
    """
    kappa, theta, sigma, rho, v0 = params
    
    # Constraints to keep parameters in reasonable ranges
    if (kappa <= 0 or theta <= 0 or sigma <= 0 or abs(rho) > 1 or v0 <= 0):
        return np.inf
    
    try:
        model_price = heston_call_price(S0, K, T, r, q, kappa, theta, sigma, rho, v0)
        return abs(model_price - market_price)
    except Exception as e:
        return np.inf

def calibrate_heston_model(subset, r=0.05, q=0.01):
    """
    Calibrate Heston model parameters for each row in the subset
    """
    # Identify the correct market price column (handle potential leading/trailing spaces)
    market_price_col = [col for col in subset.columns if 'market_price' in col.lower()][0]
    
    # Initial guess for parameters
    initial_guess = [2.0, 0.1, 0.3, -0.5, 0.1]
    
    # Bounds for parameters
    bounds = [
        (0.1, 20),    # kappa
        (0.001, 1),   # theta
        (0.01, 1),    # sigma
        (-0.99, 0.99),# rho
        (0.01, 1)     # v0
    ]
    
    # Store calibrated parameters
    calibrated_params = []
    model_prices = []
    
    # Calibrate for each row
    for index, row in subset.iterrows():
        # Optimization
        result = minimize(
            objective_function, 
            initial_guess, 
            args=(row['S0'], row['Strike'], row['Maturity_days'], r, q, row[market_price_col]),
            method='Nelder-Mead',
            bounds=bounds
        )
        
        # Calculate model price with calibrated parameters
        model_price = heston_call_price(
            row['S0'], row['Strike'], row['Maturity_days'], 
            r, q, *result.x
        )
        
        # Store calibrated parameters and model prices
        calibrated_params.append(result.x)
        model_prices.append(model_price)
    
    # Convert to DataFrame
    params_df = pd.DataFrame(
        calibrated_params, 
        columns=['kappa', 'theta', 'sigma', 'rho', 'v0']
    )
    
    # Add model prices and actual market prices
    params_df['Model_Price'] = model_prices
    params_df['Market_Price'] = subset[market_price_col]
    params_df['Price_Difference'] = params_df['Model_Price'] - params_df['Market_Price']
    
    # Visualization
    plt.figure(figsize=(20, 15))
    
    # 1. Parameter Distribution
    plt.subplot(2, 3, 1)
    params_df[['kappa', 'theta', 'sigma', 'rho', 'v0']].boxplot()
    plt.title('Parameter Distribution')
    plt.xticks(rotation=45)
    
    # 2. Price Comparison
    plt.subplot(2, 3, 2)
    plt.scatter(params_df['Market_Price'], params_df['Model_Price'])
    plt.plot([params_df['Market_Price'].min(), params_df['Market_Price'].max()], 
             [params_df['Market_Price'].min(), params_df['Market_Price'].max()], 
             'r--')
    plt.title('Market vs Model Price')
    plt.xlabel('Market Price')
    plt.ylabel('Model Price')
    
    # 3. Price Difference Histogram
    plt.subplot(2, 3, 3)
    params_df['Price_Difference'].hist(bins=20)
    plt.title('Price Difference Distribution')
    plt.xlabel('Price Difference')
    
    # 4. Correlation Heatmap of Parameters
    plt.subplot(2, 3, 4)
    sns.heatmap(params_df[['kappa', 'theta', 'sigma', 'rho', 'v0']].corr(), 
                annot=True, cmap='coolwarm', center=0)
    plt.title('Parameter Correlation')
    
    # 5. Parameters vs Moneyness
    plt.subplot(2, 3, 5)
    moneyness = subset['S0'] / subset['Strike']
    plt.scatter(moneyness, params_df['sigma'])
    plt.title('Volatility (sigma) vs Moneyness')
    plt.xlabel('Moneyness (S0/K)')
    plt.ylabel('Sigma')
    
    # 6. Parameters vs Time to Maturity
    plt.subplot(2, 3, 6)
    plt.scatter(subset['Maturity_days'], params_df['kappa'])
    plt.title('Mean Reversion (kappa) vs Maturity')
    plt.xlabel('Days to Maturity')
    plt.ylabel('Kappa')
    
    plt.tight_layout()
    plt.show()
    
    # Combine with original subset
    result_subset = pd.concat([subset, params_df], axis=1)
    
    return result_subset

# Load the data
def main():
    data = pd.read_csv('apple_data_with_volatility.csv')
    data['Maturity_days'] = data['Maturity_days']/252
    subset = data.head(10)
    
    # Calibrate the model
    calibrated_data = calibrate_heston_model(subset)
    print(calibrated_data)

if __name__ == "__main__":
    main()
```
